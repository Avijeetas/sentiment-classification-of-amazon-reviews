```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer

# Read the new dataset
df_new = pd.read_csv('1429_1.csv')

# Display the first few rows of the new dataset
# print(df_new.head())

# Visualization 1: Rating Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='reviews.rating', data=df_new)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# Visualization 2: Recommendation Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='reviews.doRecommend', data=df_new)
plt.title('Distribution of Recommendations')
plt.xlabel('Recommendation')
plt.ylabel('Count')
plt.show()

def plot_top_words(text_data, n=10):
    # Fill NaN values with an empty string
    text_data = text_data.fillna('')

    vectorizer = CountVectorizer(stop_words='english')
    X = vectorizer.fit_transform(text_data)
    word_counts = pd.DataFrame(X.sum(axis=0), columns=vectorizer.get_feature_names_out())
    top_words = word_counts.T.sort_values(by=0, ascending=False).head(n)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_words[0], y=top_words.index, palette='viridis')
    plt.title(f'Top {n} Common Words in Reviews')
    plt.xlabel('Count')
    plt.ylabel('Word')
    plt.show()

# Plot the top 10 common words in reviews
plot_top_words(df_new['reviews.text'], n=10)

```
